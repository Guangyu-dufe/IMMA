```python
import matplotlib.pyplot as plt

def plot_multiple_time_series(data, title="365 days", xlabel="time", ylabel="Value"):
    plt.figure(figsize=(20, 6))

    days = 365
    points_per_day = 288
    data_reshaped = data.reshape(days, points_per_day)

    for day in range(days):
        plt.plot(data_reshaped[day], linewidth=0.8, alpha=0.7)
    
    plt.title(title, fontsize=14)
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

plot_multiple_time_series(data)
```
