```python
import pandas as pd
import random
```

```python
df = pd.read_hdf('./ca_his_raw_2017.h5')
df2 = pd.read_hdf('./ca_his_raw_2018.h5')
df3 = pd.read_hdf('./ca_his_raw_2019.h5')
df4 = pd.read_hdf('./ca_his_raw_2020.h5')
```

```python
# 检查每个数据框中全为NaN的列
for i, df_name in enumerate(['df', 'df2', 'df3', 'df4'], 1):
    nan_counts = eval(df_name).isna().sum()
    all_nan_columns = nan_counts[nan_counts == len(eval(df_name))]
    print(f"\n{df_name}中全为NaN的列数量：{8600-len(all_nan_columns)}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

sensorid = random.randint(0, 8600)
data_2017 = df.iloc[sensorid]
data_2018 = df2.iloc[sensorid]
data_2019 = df3.iloc[sensorid]
data_2020 = df4.iloc[sensorid]
```

```python
fig, axes = plt.subplots(4, 1, figsize=(15, 6), sharex=True)

axes[0].plot(data_2017.index, data_2017.values, color='blue')

axes[1].plot(data_2018.index, data_2018.values, color='blue')

axes[2].plot(data_2019.index, data_2019.values, color='green')

axes[3].plot(data_2020.index, data_2020.values, color='red')

plt.tight_layout()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 6))

ax.plot(data_2018.index, data_2017.values, color='blue', label='2017')
ax.plot(data_2018.index, data_2018.values, color='green', label='2018')
ax.plot(data_2019.index, data_2019.values, color='orange', label='2019')
ax.plot(data_2020.index, data_2020.values, color='red', label='2020')

ax.legend()
plt.tight_layout()

plt.show()
```
